# split the dataset

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('Chance of Admit ',axis=1)
y = df['Chance of Admit ']
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
X = scaler.fit_transform(X)
```

```python
pd.DataFrame(X)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
L = LinearRegression()
```

```python
L.fit(X_train,y_train)
```

```python
print(L.score(X_test,y_test))
```

```python
y_pred = L.predict(X_test)
```

```python
sns.distplot(y_test-y_pred)
```

```python
from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(y_test, y_pred))
print('MSE:', metrics.mean_squared_error(y_test, y_pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
plt.figure(figsize=(10,10))
plt.plot(y_pred,y_test,'o')
plt.xlabel("predicted value")
plt.ylabel("actual value")
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor
R = RandomForestRegressor()
```

```python
R.fit(X_train,y_train)
```

```python
print(R.score(X_test,y_test))
```

```python
y_predR = R.predict(X_test)
```

```python
sns.distplot(y_test-y_predR)
```

```python
from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(y_test, y_predR))
print('MSE:', metrics.mean_squared_error(y_test, y_predR))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, y_predR)))
```

```python
plt.figure(figsize=(10,10))
plt.plot(y_predR,y_test,'o')
plt.xlabel("predicted value")
plt.ylabel("actual value")
plt.show()
```
